## Cleaning and data wrangling

```python
# checking for null values in the dataset

eud_df.isnull().sum().sort_values(ascending=False)
```

The fuel_unit has 180 missing values in its rows. So we have to know the importance of the feature to determine whether to drop the missing rows or fill it  with some other values.

```python
# Checking the value counts of each value in fuel_unit. mcf is the mode of this column

eud_df['fuel_unit'].value_counts()
```

```python
# checking the mode using groupby

eud_df.groupby('fuel_unit')['fuel_unit'].count()
```

Since mcf is the highest occuring value
 we are going to fill the missing values with mcf.
 Note this might not be the case in so many projects
 some may require you completely drop the null value rows
 To drop values: **df.dropna()**

```python
# FIlling the missing values

eud_df['fuel_unit'] = eud_df['fuel_unit'].fillna('mcf')
```

```python
# Verify if the missing values has been filled

eud_df.isnull().sum()
```

+ As we can see there are no more columns with missing values. So we have been able to solve the missing value problem

### Merging in Pandas

```python
eud_df1 = eud_df.iloc[0:19000].reset_index(drop=True)
```

```python
eud_df2 = eud_df.iloc[19000:].reset_index(drop=True)
```

```python
# Check that the length of both dataframes sum to the expected lengths

assert len(eud_df) == (len(eud_df1)+len(eud_df2))
```

#### There are merging methods in pandas merge() function. They are right,left,outer and inner

Lets explore some of them

```python
# An inner merge will lose rows that do not match in the dataframe

eg1 = pd.merge(eud_df1,eud_df2,how='inner')
```

```python
# outer merge returns all rows in both dataframes

eg2 = pd.merge(eud_df1,eud_df2,how='outer')
```

```python
# removes rows from the right dataframe that do not have a match with the left
# and keeps all rows from the left

eg3 = pd.merge(eud_df1,eud_df2,how='left')
```

```python
# removes rows from the left dataframe that do not have a match with the right
# and keeps all rows from the right

eg4 = pd.merge(eud_df1,eud_df2,how='right')
```
